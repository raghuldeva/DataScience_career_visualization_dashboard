```python
# STEP 1

import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import re
import time

import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# STEP 2

link = []
for pg_no in range(1, 16):
    url = "https://www.simplyhired.co.in/search?q=data+science&pn="+str(pg_no)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    time.sleep(2)
    for t in soup.findAll('a', attrs = {'href': re.compile('/job/')}):
        q = t.get('href')
        url1 = "https://www.simplyhired.co.in"+ q
        #q = q.split('\n')
        link.append(url1)

#print(link)
```

```python
len(link)
```

```python
# STEP 3

job_title_list = []
company_name_location_list = []
working_time_list = []
job_age_list = []
qualification_list = []
benefits_list = []
description_list = []

count = 0
for i in link:
    url = i
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    time.sleep(2)
    count +=1
    print(url, count)
    # Job Title
    for job_title in soup.find_all('div', {'class':'viewjob-jobTitle h2'}):
        job_title_list.append(job_title.get_text())
        print(job_title.get_text())
        
    # Company name, location
    for companyname_location in soup.find_all('div', {'class':'viewjob-labelWithIcon'}):
        company_name_location_list.append(companyname_location.get_text())
        print(companyname_location.get_text())
        
    # Working Time
    if soup.find_all('span', {'class':'viewjob-labelWithIcon viewjob-jobType'}) == []:
        working_time_list.append(np.NaN)
        print(np.NaN)
    else:
        for working_time in soup.find_all('span', {'class':'viewjob-labelWithIcon viewjob-jobType'}):
            working_time_list.append(working_time.get_text())
            print(working_time.get_text())
    
    # Columns which have job age
    if soup.find_all('span', {'class':'viewjob-labelWithIcon viewjob-age'}) == []:
        job_age_list.append(np.NaN)
        print(np.NaN)
    else:
        for job_age in soup.find_all('span', {'class':'viewjob-labelWithIcon viewjob-age'}):
            job_age_list.append(job_age.get_text())
            print(job_age.get_text())
            
    # Columns which have Qualification
    temp_list = []    
    for qualification in soup.find_all('li', {'class':'viewjob-qualification'}):
        temp_list.append(qualification.get_text())
    print(temp_list)
    qualification_list.append(temp_list)
    
    # Job benefits
    if soup.find_all('li', {'class':'viewjob-benefit'}) == []:
        benefits_list.append("Not mentioned")
        print("Not mentioned")
    else:
        benefits_list_temp = []
        for benefits in soup.find_all('li', {'class':'viewjob-benefit'}):
            temp = benefits.get_text().split('\n')
            for i in temp:
                benefits_list_temp.append(i)
        print(benefits_list_temp)
        benefits_list.append(benefits_list_temp)
    
    # Job Decsription
    for description in soup.find_all('div', {'class':'p'}):
        description_list.append(description.get_text())
        print(description.get_text())
```

```python
print("Job Title : ", len(job_title_list))
print("Company name location : ", len(company_name_location_list))
print("Working time : ", len(working_time_list))
print("Job age : ", len(job_age_list))
print("Qualification : ", len(qualification_list))
print("Benefits : ", len(benefits_list))
print("Description : ", len(description_list))
```

```python
# Splitting company name and location
company_name_list = []
for i in range(0, len(company_name_location_list), 2):
    company_name_list.append(company_name_location_list[i].replace(' - ', ''))
    
company_location_list = []
for i in range(1, len(company_name_location_list), 2):
    company_location_list.append(company_name_location_list[i])
```

```python
pd.set_option("display.max_colwidth", -1)
df = pd.DataFrame([job_title_list, company_name_list, company_location_list, working_time_list, job_age_list, qualification_list, benefits_list, description_list]).transpose()
```

```python
df.columns=['Job_title', 'company_name', 'company_location', 'job_type', 'posted_date', 'qualification', 'benefits', 'description']
```

```python
import pandas as pd
df = pd.read_csv('simply_hired_data.csv')
```

```python
company_name_list = [i.replace(' - ', '') for i in df['company_name']]
```

```python
job_title_list = [i for i in df['Job_title']]
```

```python
glassdoor_url = []
for i, j in zip(company_name_list, job_title_list):
    s = ""
    s = i.replace(' ', '+') + str('+') + j.replace(' ', '+')
    glassdoor_url.append("https://www.glassdoor.co.in/Job/jobs.htm?suggestCount=0&suggestChosen=false&clickSource=searchBtn&typedKeyword="+str(s)+"&typedLocation=&locT=&locId=&jobType=&context=Jobs&sc.keyword="+str(s)+"&dropdown=0")
```

```python
careerbliss_url = []
for i in company_name_list:
    name = i.split()[0]
    s = ""
    s = "https://www.careerbliss.com/" + str(name)+ "/reviews/"
    careerbliss_url.append(s)
```

```python
average_rating_list = []
company_culture = []
growth_opportunities = []
people_you_work_with = []
person_you_work_for = []
rewards_you_receive = []
suppport_you_get = []
way_you_work = []
work_setting = []

count = 0
for url in careerbliss_url:
    count += 1
    r = requests.get(url)
    print(url, count, r.status_code)
    soup = BeautifulSoup(r.text, 'html.parser')
    time.sleep(2)

    if r.status_code != 200:
        average_rating_list.append(np.NaN)
        company_culture.append(np.NaN)
        growth_opportunities.append(np.NaN)
        people_you_work_with.append(np.NaN)
        person_you_work_for.append(np.NaN)
        rewards_you_receive.append(np.NaN)
        suppport_you_get.append(np.NaN)
        way_you_work.append(np.NaN)
        work_setting.append(np.NaN)
    else:
        if soup.find('div', {'class':'header7'}) == None:
            average_rating_list.append(np.NaN)
        else:
            for average_rating in soup.find('div', {'class':'header7'}):
                average_rating_list.append(average_rating.get_text())

        rating = []
        if soup.find_all('span', {'class':'span2 header8'}) == []:
            rating = [np.NaN] * 8
        else:
            for category_rating in soup.find_all('span', {'class':'span2 header8'}):
                for i in category_rating.get_text().split():
                    rating.append(i)

        for i in range(0, len(rating)):
            if i == 0:
                company_culture.append(rating[i])
            elif i == 1:
                growth_opportunities.append(rating[i])
            elif i == 2:
                people_you_work_with.append(rating[i])
            elif i == 3:
                person_you_work_for.append(rating[i])
            elif i == 4:
                rewards_you_receive.append(rating[i])
            elif i == 5:
                suppport_you_get.append(rating[i])
            elif i == 6:
                way_you_work.append(rating[i])
            elif i == 7:
                work_setting.append(rating[i])   
    
```

```python
print(len(average_rating_list))
print(len(company_culture))
print(len(growth_opportunities))
print(len(people_you_work_with))
print(len(person_you_work_for))
print(len(rewards_you_receive))
print(len(suppport_you_get))
print(len(way_you_work))
print(len(work_setting))
```

```python
careerbliss_url_about = []
for i in company_name_list:
    name = i.split()[0]
    s = ""
    s = "https://www.careerbliss.com/" + str(name)+ "/"
    careerbliss_url_about.append(s)
```

```python
reviews = []

count = 0
for url in careerbliss_url_about:
    count += 1
    r = requests.get(url)
    print(url, count, r.status_code)
    soup = BeautifulSoup(r.text, 'html.parser')

    if r.status_code != 200:
        print(np.NaN)
        reviews.append(np.NaN)
    else:
        if soup.find('i', {"class":"reviews-carousel-text"}) == None:
            reviews.append(np.NaN)
            print(np.NaN)
        else:
            for i in soup.find('i', {"class":"reviews-carousel-text"}):
                reviews.append(i.get_text())
                print(i.get_text())
```
